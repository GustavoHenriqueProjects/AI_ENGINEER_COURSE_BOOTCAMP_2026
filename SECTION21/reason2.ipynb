```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

en_stopwords = stopwords.words('english')
print(en_stopwords)
```

```python
setence = "I'm going to the movies tonight"
setence_no_stopwords = ' '.join([word for word in setence.split() if word not in en_stopwords])
print(setence_no_stopwords)

en_stopwords.remove('to')
en_stopwords.append('tonight')
setence_no_stopwords = ' '.join([word for word in setence.split() if word not in en_stopwords])
print(setence_no_stopwords)
```
