```python
import re

my_fonder = "C:/Users/gusta/OneDrive/Documentos/GitHub/AI_ENGINEER_COURSE_BOOTCAMP_2026/SECTION21/aula_02_regex.ipynb"

## Com caractere especial
my_fonder = r"C:\Users\gusta\OneDrive\Documentos\GitHub\AI_ENGINEER_COURSE_BOOTCAMP_2026\SECTION21\aula_02_regex.ipynb"

print(my_fonder)

## Utilizando biblioteca re

result_search = re.search("cook", r"I like to cook")
print(result_search)

result_search2 = re.search("cook", r"I like to play football")
print(result_search2)

## Utilizando método re.sub para substituir o texto
string = r"My sister is Fernanda"
new_string = re.sub("Fernanda", "Fernanda Regina", string)
print(new_string)

customers_reviews = [
    "I love the product, but the service was not good.",
    "The product is great, but the service is bad",
    "I love the product, but the service was not good",
    "The product is great, but the service is bad",
    "I love the product, but the service was not good and the Fernanda was not good",
    "My product is great, but the service is bad, my sister Fernanda is not good",
    "I needed a new product, but the service is bad",
]

fernanda_review = []
pattern_to_find = r"Fernanda"

for line in customers_reviews:
    if (re.search(pattern_to_find, line)):
        fernanda_review.append(line)

print(fernanda_review)

## Busca avaliações que começam com "My"
a_reviews = []
pattern_to_find = r"^My"

for line in customers_reviews:
    if (re.search(pattern_to_find, line)):
        a_reviews.append(line)

print(a_reviews)

## Busca avaliações que terminam com "good"
good_reviews = []
pattern_to_find = r"good$"

for line in customers_reviews:
    if (re.search(pattern_to_find, line)):
        good_reviews.append(line)

print(good_reviews)

## Busca avaliações utilizando o operador OR

needwant_reviews = []
pattern_to_find = r"(need|was)ed"

for line in customers_reviews:
    if (re.search(pattern_to_find, line)):
        needwant_reviews.append(line)

print(needwant_reviews)

## Removendo pontuações ou outros caracteres especiais no texto
no_punct_pontuation = []
pattern_to_find = r"[^\w\s]"

for line in customers_reviews:
    if (re.search(pattern_to_find, line)):
        no_punct_pontuation.append(re.sub(pattern_to_find, "", line))
print(no_punct_pontuation)
```
