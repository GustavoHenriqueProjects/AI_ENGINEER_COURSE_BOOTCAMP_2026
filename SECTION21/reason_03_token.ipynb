```python
import nltk
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
sentences = "Her cat's name is Fluffy. her dog's name is Spot."

##Divide cada frase em uma lista ou token
sent_tokenize(sentences)

##Divide cada palavra em uma lista ou token
sentence = "her cat's name is Fluffy."
word_tokenize(sentence)

sentence_2 = "her cat's name is Fluffy and her dog's name is Spot."
word_tokenize(sentence_2)
```
