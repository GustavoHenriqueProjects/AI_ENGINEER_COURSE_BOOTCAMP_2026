```python
import nltk
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize, sent_tokenize

from nltk.stem import PorterStemmer

##Objetivo: reduzir as palavras aos seus radicais ex: running, runner, runs, ran, etc.
ps = PorterStemmer()
connect_tokens = ['connecting', 'connected', 'connection', 'connections']
for t in connect_tokens:
    print(t, ": ", ps.stem(t))
    
learn_tokens = ['learning', 'learned', 'learner', 'learns']
for t in learn_tokens:
    print(t, ": ", ps.stem(t))
    
likes_tokens = ['likes', 'better', 'worse', 'likings']
for t in likes_tokens:
    print(t, ": ", ps.stem(t))
    




```
